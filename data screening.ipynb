```python
import pandas as pd
origin_data=pd.read_excel(r'D:/JCE/ML_data.xlsx')
import math
from scipy.integrate import quad
import numpy as np
correlation_data=origin_data.copy()
finial_data=pd.DataFrame()
origin_data
for i in range(int(len(origin_data)/6)):
    for j in range(6):
        k=j+6*i
        if correlation_data.iloc[k,0]<=366.63:
            correlation_data.loc[k,'pethanol*']=10**(5.24677-1598.673/(correlation_data.iloc[k,0]-46.424))
        else: correlation_data.loc[k,'pethanol*']=10**(4.92531-1432.526/(correlation_data.iloc[k,0]-61.819))
        correlation_data.loc[k,'pn_propanol*']=10**(4.87601-1441.629/(correlation_data.iloc[k,0]-74.299))
        correlation_data.loc[k,'γethanol']=1*(1-correlation_data.loc[k,'Vapor phase concentration'])/((correlation_data.loc[k,'pethanol*']*(1-correlation_data.loc[k,'Liquid phase concentration'])))
        correlation_data.loc[k,'γn_propanol']=1*correlation_data.loc[k,'Vapor phase concentration']/((correlation_data.loc[k,'pn_propanol*']*correlation_data.loc[k,'Liquid phase concentration']))
        correlation_data.loc[k,'ln(γn_propanol/γethanol)']=math.log(correlation_data.loc[k,'γn_propanol']/correlation_data.loc[k,'γethanol'])
for g in range(int(len(origin_data)/6)):
    x=correlation_data.loc[g*6:g*6+5,'Liquid phase concentration']
    y=correlation_data.loc[g*6:g*6+5,'ln(γn_propanol/γethanol)']
    f=np.polyfit(x,y,2)
    polynomial = np.poly1d(f)
    result=quad(polynomial, 0, 1)[0]
    def abs_polynomial(x):
        return np.abs(polynomial(x))
    abs_result=quad(abs_polynomial, 0, 1)[0]
    D=abs(result/abs_result)
    if D<=0.05:
        finial_data=pd.concat([finial_data,correlation_data.iloc[g*6:g*6+6]],axis=0)
model_data=finial_data.reset_index(drop=True)
model_data
```
